## Strings

```python
a = 'hello' # String literals can use single quotes
b = "world" # or double quotes; it does not matter.
print(len(a))   # String length; prints "5"
ab = a + ' ' + b # String concatenation
print(ab)    # prints "hello world"
print(a.capitalize())  # Capitalize a string; prints "Hello"
print(a.upper())  # Convert a string to uppercase; prints "HELLO"
print(a.replace('l', 'x'))  # Create a string from a by replacing all instances of one substring with another;
                      # prints "hexxo"
print('     hello       '.strip())  # Strip leading and trailing whitespace; prints "hello"
```

## Lists

A list is the Python equivalent of an array, but is resizeable and can contain elements of different types:


```python
x = [4, 1, 'b']   # Create a list [4, 1, 'b']
print(x[0], x[2])  # Prints "4 ´b´"
print(x[-1])     # Negative indices count from the end of the list; prints "b"
a = x.pop()      # Remove and return the last element of the list
print(a)
x.sort() # Sort x
print(x)
x.append('ola')  # Add a new element to th´end of the list
print(x)   

# Slicing
y = list(range(11)) # Create a list containing 0, 1, ..., 10
print(y) # Print list
print(y[:4]) # print the elements of y from the start until index 4 (exclusive)
print(y[:-1]) # print all elements, except the last one
print(y[1::2])# Show the odd numbers
print(y[::-1])# Show y in the reverse order

# List comprehensions
# - Create a list and take it to the power of 3, using for
z = list(range(21))
z3 = []
for value in z:
    z3.append(value**3)
print(z3)
# - using list comprehensions
z3 = [value**3 for value in z]
print(z3)

# - using numpy
import numpy as np
print(np.power(z, 3))

# Reference
# ´Copy the reference of a list:
w = z
w[0] = 400
print(w)
print(z)

# Copy the content of a list:
w2 = z.copy()
w2[0] = 500
print(w2)
print(z)
```

```python
import numpy as np

# Arrays
# A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.
# - Create a numpy array and print its dimensions
a = np.array([[1,2,3],[4,5,6]])
print(a.shape)
# - Show elements in [0, 1]
print(a[0,1])
# - Show a submatrix
print(a[:,1:])
# - Boolean array indexing
a[a < 5] = 3
print(a)
# - Compute the mean of each row
print(np.mean(a, 1))
```

## NetworkX

Library for manipulating and analyzing networks

```python
import networkx as nx

# Create an empty graph
G = nx.Graph()

# Add nodes and edges
G.add_nodes_from(['a', 'b', 'c'])
G.add_edge('a', 'b')

# Print graph information
print(nx.info(G))
# Draw the resulting graph

nx.draw(G, with_labels=True)
```

## Exercise 1

* Create a random binary symmetric matrix A of dimensions 143 x 143, in which all elements of the main diagonal are zero
* Pick at random 6 positions of the network containing ones and replace them by zero
* Remove all rows and columns containing only zeros, if there is any
* Create a graph G_1 in which each row (or column) of A corresponds to a node, and the 1s indicates edges. (tip: use the function nx.from_numpy_matrix)
* Draw the resulting graph

```python
# Write your code here
# Set seed to repeat the same experiment
np.random.seed(0)
# Generate random binary matrix
A = np.random.randint(2, size=(143, 143))
# Create a symmetric version of A with diagonal zero
ones = np.where(np.triu(A) == 1)
# Remove connections chosen at random
edges = np.random.randint(np.shape(ones)[1], size=6)
A[ones[0][edges], ones[1][edges]] = 0
A = np.triu(A, 1) + np.triu(A, 1).T
to_remove = np.where(np.sum(A, axis=1)==0)
# Remove vertices with degree equal to zero
if to_remove[0].size > 0:
    A = np.delete(A, to_remove[0], 1)
    A = np.delete(A, to_remove[0], 0)
# Build a graph with the resulting adjacency matrix
G_1 = nx.from_numpy_array(A)
print(nx.info(G_1))
nx.draw(G_1)
```

```python
# Write your code here
import random
import matplotlib.pyplot as plt

random.seed(0)
G_2 = nx.read_edgelist("ia-enron-only.edges")
G_2 = nx.convert_node_labels_to_integers(G_2)
print(nx.info(G_2))
G_2.remove_edges_from(random.sample(list(G_2.edges), 6))
degrees = np.array(G_2.degree)
to_remove = np.where(degrees[:,1] == 0)
if len(to_remove[0]) > 0:
    G_2.remove_nodes_from(degrees[to_remove[0],0])
print(nx.info(G_2))
# Force the networks to have the same number of nodes
if (G_2.number_of_nodes() != G_1.number_of_nodes()):
    nodes = set(G_1.nodes).difference(set(G_2.nodes))
    G_1.remove_nodes_from(nodes)

print(nx.info(G_1))
nx.draw(G_2)

# Remove edges from G_1 so that it has the same number of edges than G_2
G_1.remove_edges_from(random.sample(list(G_1.edges), 
                                    G_1.number_of_edges()-G_2.number_of_edges()))
print(nx.info(G_1))

```

## Exercise 3

* Assume that G_1 and G_2 are graphs on the same set of vertices. Then build a graph G_3 containing the union of edges from G_1 and G_2
* What is the intersection between the two sets of edges?
* Save the list of edges from G_3 into a txt file
* Draw G_3

```python

# Generate a graph containing the union between G_1 and G_2
G_3 = nx.from_edgelist(set(G_1.edges).union(set(G_2.edges)))
print(nx.info(G_3))

# Show the intersection
intersec = set(G_1.edges).intersection(set(G_2.edges))
print(len(intersec))
print(intersec)

nx.write_edgelist(G_3, "edges_G_3.txt")

nx.draw(G_3)
```
