```python
import pandas as pd
import sys, os

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
DATA_PATH="../data/"
```

```python
dbs = ["cedrinho", "guaribu", "ipe", "jatoba", "mandioqueiro", "virola"]
```

```python
db_df = pd.DataFrame()
```

```python
for db in dbs:
    table_df = pd.read_csv(f"{DATA_PATH}{db}.csv", low_memory=False, sep=",")
    db_df = pd.concat([db_df, table_df], ignore_index=True)
```

### Graph by Id only

```python
id_graph_df = db_df[["OrigemID", "DestinoID"]].reset_index(drop=True).set_index(["OrigemID"]).drop_duplicates()
id_graph_df.reset_index(inplace=True, drop=True)
id_graph_df.sort_values(by="OrigemID", inplace=True)
id_graph_df
```

```python
origem_nodes = id_graph_df[["OrigemID"]].drop_duplicates()
origem_nodes.rename(columns={"OrigemID":"ID"}, inplace=True)
origem_nodes
```

```python
desino_nodes = id_graph_df.reset_index()[["DestinoID"]].drop_duplicates()
desino_nodes.rename(columns={"DestinoID":"ID"}, inplace=True)
desino_nodes
```

```python
total_nodes = pd.concat([origem_nodes, desino_nodes], ignore_index=True).drop_duplicates()
total_nodes
```

### Graph by Complex ID

```python
graph_df = db_df[["NomeOrigem", "OrigemID","UFOrigem", "MunOrigem", "NomeDestino", "DestinoID", "UFDestino", "MunDestino"]].reset_index(drop=True).drop_duplicates()
graph_df.reset_index(inplace=True, drop=True)
graph_df.sort_values(by="OrigemID", inplace=True)
graph_df
```

```python
origem_nodes = graph_df[["NomeOrigem", "OrigemID","UFOrigem", "MunOrigem"]].drop_duplicates()
origem_nodes.rename(columns={"NomeOrigem":"Name", "OrigemID":"ID","UFOrigem":"UF", "MunOrigem":"Mun"}, inplace=True)
origem_nodes
```

```python
desino_nodes = graph_df.reset_index()[["NomeDestino","DestinoID","UFDestino", "MunDestino"]].drop_duplicates()
desino_nodes.rename(columns={"NomeDestino":"Name", "DestinoID":"ID","UFDestino":"UF", "MunDestino":"Mun"}, inplace=True)
desino_nodes
```

```python
total_nodes = pd.concat([origem_nodes, desino_nodes], ignore_index=True).drop_duplicates()
total_nodes
```
